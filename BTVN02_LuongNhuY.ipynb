```python
# import thư viện
# 22C01018 - Lê Hoài Thanh
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
import math as math
```

```python
def Triangular_KDE(samples, bandwidth, x):
  n = samples.size

  sum = 0
  for i in range(n):
    u = (samples[i] - x)/bandwidth
    if (np.abs(u) <= 1):      
      sum += (1 - np.abs(u))

  kernel_f = 1/(n*bandwidth) * sum
  
  return kernel_f
```

```python
# Tạo mẫu dữ liệu Norm
norm_sample = sp.stats.norm.rvs(0,1,size=1000)

x = []
y1 = []
y2 = []
y3 = []

# [-5, 5] / 1000 lần
inc = -5
for i in range(1000):
  inc += 10/1000

  x.append(inc)

  value1 = Triangular_KDE(norm_sample, 0.5, inc)
  value2 = Triangular_KDE(norm_sample, 2.0, inc)
  value3 = Triangular_KDE(norm_sample, 5.0, inc)
  y1.append(value1)
  y2.append(value2)
  y3.append(value3)
```

```python
# Biểu diễn trên đồ thị
fig = plt.figure(figsize=(15,8))
plt.hist(norm_sample,80,density = True, color='b')
plt.plot(x,y1,color = 'r')
plt.plot(x,y2,color = 'y')
plt.plot(x,y3,color = 'g')
plt.show()
```

```python
def Uniform_KDE(samples, bandwidth, x):
  n = samples.size

  sum = 0
  for i in range(n):
    u = (samples[i] - x)/bandwidth
    if (np.abs(u) <= 1):      
      sum += (1/2)

  kernel_f = 1/(n*bandwidth) * sum
  
  return kernel_f
```

```python
# Uniform Kernel
x = []
y1 = []
y2 = []
y3 = []

# [-5, 5] / 1000 lần
inc = -5
for i in range(1000):
  inc += 10/1000

  x.append(inc)

  value1 = Uniform_KDE(norm_sample, 0.8, inc)
  value2 = Uniform_KDE(norm_sample, 2.0, inc)
  value3 = Uniform_KDE(norm_sample, 5.0, inc)
  y1.append(value1)
  y2.append(value2)
  y3.append(value3)
```

```python
# Biểu diễn trên đồ thị
fig = plt.figure(figsize=(15,8))
plt.hist(norm_sample,bins=80,density = True, color='b')
plt.plot(x,y1,color = 'r')
plt.plot(x,y2,color = 'y')
plt.plot(x,y3,color = 'g')
plt.show()
```

```python
def Parabolic_KDE(samples, bandwidth, x):
  n = samples.size

  sum = 0
  for i in range(n):
    u = (samples[i] - x)/bandwidth
    if (np.abs(u) <= 1):      
      sum += 3/4*(1-u**2)

  kernel_f = 1/(n*bandwidth) * sum
  
  return kernel_f
```

```python
def Biweight_KDE(samples, bandwidth, x):
  n = samples.size

  sum = 0
  for i in range(n):
    u = (samples[i] - x)/bandwidth
    if (np.abs(u) <= 1):      
      sum += 15/16*((1-u**2)**2)

  kernel_f = 1/(n*bandwidth) * sum
  
  return kernel_f
```

```python
def Gaussian_KDE(samples, bandwidth, x):
  n = samples.size

  sum = 0
  for i in range(n):
    u = (samples[i] - x)/bandwidth
    sum += 1/(math.sqrt(2*np.pi))*math.exp(-1/2*(u**2))

  kernel_f = 1/(n*bandwidth) * sum
  
  return kernel_f
```

```python
# Uniform Kernel
x = []
y1 = []
y2 = []
y3 = []

# [-5, 5] / 1000 lần
inc = -5
for i in range(1000):
  inc += 10/1000

  x.append(inc)

  value1 = Gaussian_KDE(norm_sample, 0.05, inc)
  value2 = Gaussian_KDE(norm_sample, 2.0, inc)
  value3 = Gaussian_KDE(norm_sample, 5.0, inc)
  y1.append(value1)
  y2.append(value2)
  y3.append(value3)
```

```python
# Biểu diễn trên đồ thị
fig = plt.figure(figsize=(15,8))
plt.hist(norm_sample,bins=80,density = True, color='b')
plt.plot(x,y1,color = 'r')
plt.plot(x,y2,color = 'y')
plt.plot(x,y3,color = 'g')
plt.show()
```

```python
def Silverman_KDE(samples, bandwidth, x):
  n = samples.size

  sum = 0
  for i in range(n):
    u = (samples[i] - x)/bandwidth
    sum += 1/2*math.exp(-(np.abs(u))/(np.sqrt(2)))*np.sin(np.abs(u)/np.sqrt(2)+np.pi/4)

  kernel_f = 1/(n*bandwidth) * sum
  
  return kernel_f
```

```python
# Uniform Kernel
x = []
y1 = []
y2 = []
y3 = []

# [-5, 5] / 1000 lần
inc = -5
for i in range(1000):
  inc += 10/1000

  x.append(inc)

  value1 = Silverman_KDE(norm_sample, 0.1, inc)
  value2 = Silverman_KDE(norm_sample, 2.0, inc)
  value3 = Silverman_KDE(norm_sample, 5.0, inc)
  y1.append(value1)
  y2.append(value2)
  y3.append(value3)
```

```python
# Biểu diễn trên đồ thị
fig = plt.figure(figsize=(15,8))
plt.hist(norm_sample,bins=80,density = True, color='b')
plt.plot(x,y1,color = 'r')
plt.plot(x,y2,color = 'y')
plt.plot(x,y3,color = 'g')
plt.show()
```
